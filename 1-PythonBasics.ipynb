```python
# Lists
demo_list = [1, 4, 2, 5, 3]
# List size
print("List size: ", len(demo_list))
# Add an element at the end
demo_list.append(6)
print("Append 6:", demo_list)
# Remove an element
demo_list.remove(6)
print("Remove 6:", demo_list)
# Concatenate lists
demo_list += [7, 6]
print("Concatenate with [7,6]:", demo_list)
# Sort list
demo_list = sorted(demo_list)
print("Sorted list:", demo_list)
# Slicing
# Indices go from 0 to len(demo_list) - 1
# You can create subsets of a list
# From the beginning to index 3.
print("Beginning to index 3 subset:", demo_list[:4]) # Extracts index 0, 1, 2, 3. Index 4 is excluded
# From index 3 to the end.
print("From index 3 to end subset:", demo_list[3:])
# Between index 1 and 3
print("From index 1 to index 3 subset:", demo_list[1:4])
```

## 5. [Loops](https://www.geeksforgeeks.org/python/loops-in-python/)

### 5.1 For loop
Best when the number of iterations is known beforehand. The size of the object size you're iterating over shouldn't change during the loop.


```python
vector1 = [1, 2, 3, 4, 5]
vector2 = [6, 7, 8, 9, 10]
# Iterate over the list values
print("By value:")
for value in vector1:
    print(value)
print("\n")
# Iterate over the list using index
for i in range(len(vector1)):
    print(f"Index: {i} / Value: {vector1[i]}")
print("\n")
# Iterate over two vectors
for value1, value2 in zip(vector1, vector2):
    print(f"Value1: {value1} / Value2: {value2}")
```

### 5.2 While loop
Best when looping depends on a dynamic condition. Risks infinite loops if the exit condition is never satisfied.

```python
vector1 = [1, 2, 3, 4, 5]
index = 0
while len(vector1) > 0: # Stops when there are no more items left in the list
    value = vector1.pop(0) # Remove first element at each iteration
    index += 1
    print(f"Index: {index} / Value: {value} / Vector size: {len(vector1)}")
```

```python
def binary_search(arr: list[int], a_target: int) -> bool:
    low, high = 0, len(arr) - 1
    while low <= high:
        m = low + (high - low) // 2
        # Check if x is present at mid
        if arr[m] == a_target:
            return True

        # If x is greater, ignore left half
        elif arr[m] < a_target:
            low = m + 1

        # If x is smaller, ignore right half
        else:
            high = m - 1
    return False
import time
start = time.time()
binary_search(vector, target)
print("Target found in", time.time() - start, "seconds.")
```

## 6. [Exception handling](https://www.geeksforgeeks.org/python/python-exception-handling/)
